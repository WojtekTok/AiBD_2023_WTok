Zad 3
Wyrysowanie wykresów dla funkcji dla 3 różnych zakresów

```python
f = lambda x: x**2 + 5 # funkcja
x = np.linspace(-1,1,100) # zakres
y = f(x)
fig1 = plt.figure()
plt.plot(x, y, label='x**2+5')
plt.title('x**2+5 dla -1<x<1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
x = np.linspace(-6,6,600)
y = f(x)
fig2 = plt.figure()
plt.plot(x, y, label='x**2+5')
plt.title('x**2+5 dla -6<x<6')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
x = np.linspace(0,5,100)
y = f(x)
fig3 = plt.figure()
plt.plot(x, y, label='x**2+5')
plt.title('x**2+5 dla -1<x<1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
