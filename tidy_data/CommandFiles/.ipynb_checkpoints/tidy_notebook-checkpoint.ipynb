```python
import pandas as pd
```

```python
DIR = "../AnalysisData/"
DIR_o = "../OriginalData/"

tb = pd.read_csv(f"{DIR_o}tb.csv")
```

```python
tb
```

```python
tb = tb.dropna(thresh=3) #usuwam wiersze, gdzie jest mniej niż 3 wartości różne od Nan (iso i year zawsze będą różne, chcę aby jedna)
tb
```

liczba wierszy zmniejszona o prawie połowę

```python
# zmiana nazw kolumn tak, aby było jasne o kogo chodzi w danej kolumnie
tb = tb.rename(columns={'new_sp_m04': 'male, 0-4 years', 'new_sp_m514': 'male, 5-14 years', 'new_sp_m014': 'male, 0-14 years',
                       'new_sp_m1524': 'male, 15-24 years', 'new_sp_m2534': 'male, 25-34 years',
                        'new_sp_m3544': 'male, 35-44 years', 'new_sp_m4554': 'male, 45-54 years', 
                        'new_sp_m5564': 'male, 55-64 years', 'new_sp_m65': 'male, 65+ years', 'new_sp_mu': 'male, u',
                        'new_sp_f04': 'female, 0-4 years', 'new_sp_f514': 'female, 5-14 years', 'new_sp_f014': 'female, 0-14 years',
                        'new_sp_f1524': 'female, 15-24 years', 'new_sp_f2534': 'female, 25-34 years', 
                        'new_sp_f3544': 'female, 35-44 years', 'new_sp_f4554': 'female, 45-54 years',
                        'new_sp_f5564': 'female, 55-64 years', 'new_sp_f65': 'female, 65+ years', 'new_sp_fu': 'female, u'
                       })
```

```python
# Zamiana typu na int, żeby nie było nic po przecinku
for col, row in tb.iteritems():
    if col != 'iso2' and col != 'year':
        tb[col] = tb[col].astype('Int64')
tb
```

```python
tb.to_csv(path_or_buf=f"{DIR}tb.csv", index=False)
```
